```python
#import libraries!
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pandas as pd
import requests

pd.options.mode.chained_assignment = None
pd.set_option('display.max_columns', None)

%matplotlib inline

import time
from sklearn import linear_model, metrics, model_selection, feature_selection, cross_validation



def get_linear_model_metrics(X, y, algo):
    # get the pvalue of X given y. Ignore f-stat for now.
    pvals = feature_selection.f_regression(X, y)[1]
    # start with an empty linear regression object
    # .fit() runs the linear regression function on X and y
    algo.fit(X,y)
    residuals = (y-algo.predict(X)).values

    # print the necessary values
    print 'P Values:', pvals
    print 'Coefficients:', algo.coef_
    print 'y-intercept:', algo.intercept_
    print 'R-Squared:', algo.score(X,y)
    plt.figure()
    plt.hist(residuals)
    # keep the model
    return algo
```

```python
#Import df_tiny ()
df_tiny = pd.read_csv("LIHTC_Properties_wGeocode_Census.csv",low_memory=False)
```

### Add a column that compares each census tract income to it's state median income.

```python
#import the state dataset (the same API as before was run on this in a seperate notebook, then saved to a CSV)
df_states = pd.read_csv("States_with_Income.csv",low_memory=False)
df_states['PROJ_ST'] = df_states['Abbrev']
df_states['Code'] = df_states['Code'].astype(str).str.strip().astype(int)
df_states['StateCode'] = df_states.Code.map("{:02}".format)
print len(df_states)
df_states.head(10)
```

### Refine the data set and create dummy variables, as well as other calculations
#### We're predicting the ACS Median Income in 2014....NOT the percent change as originally thought.

There is still some data cleanup to be done - especially since I wasn't able to determine the variables used in the Linear Regression until I could see some R-squareds and MSE's.

This was also where I could go back substitute, refine, and add additional variables to my model.


#### My Final Predictor Variables

-Income in 2010 (based on the American Community Survey)

-Number of Low-Income units added to the Census Tract

-Difficult Development Area (areas that have high construction costs)

-Targeted Population (1 indicated if it's a elderly/homeless/disable project, 0 indicates if it's a general family property)

Note: I had tested a regression using a comparison against the states' median income - i.e. is this census tract above or below the state median income? This would indicate that a project was placed in a rich neighborhood. However, that was found to be very collinear to Income in 2010.

```python
df_tiny =  df_tiny[(df_tiny.YR_PIS == 2006)|(df_tiny.YR_PIS == 2007)|(df_tiny.YR_PIS == 2008)|(df_tiny.YR_PIS == 2009)|(df_tiny.YR_PIS == 2010) | (df_tiny.YR_PIS == 2011)|(df_tiny.YR_PIS == 2012)|(df_tiny.YR_PIS == 2013)|(df_tiny.YR_PIS == 2014)|(df_tiny.YR_PIS == 2015)]
#df_tiny =  df_tiny[(df_tiny.YR_PIS == 2006)]
df_tiny = df_tiny[np.isfinite(df_tiny['MHI_2014'])]
df_tiny = df_tiny[np.isfinite(df_tiny['MHI_2010'])]
df_tiny = df_tiny[np.isfinite(df_tiny['HH_2010'])]
df_tiny = df_tiny[np.isfinite(df_tiny['N_UNITS'])]
df_tiny = df_tiny[np.isfinite(df_tiny['LI_UNITS'])]

#creating income growth 
df_tiny['10_14_Growth'] = df_tiny['MHI_2014'].astype(int) - df_tiny['MHI_2010'].astype(int)
df_tiny['PercentGrowth'] = (df_tiny['MHI_2014'].astype(int) - df_tiny['MHI_2010'].astype(int)) / df_tiny['MHI_2010'].astype(int)


df_tiny['Code'] = df_tiny['STATE2KX'].astype(str).str.strip().astype(int)
df_tiny['StateCode'] = df_tiny.Code.map("{:02}".format)
df_tiny['N_UNITS'] = df_tiny['N_UNITS'].astype(float)
df_tiny['TRGT_FAM'] = df_tiny['TRGT_FAM'].astype(int)

print "unique values in  target family: " + str(df_tiny.TRGT_FAM.unique())
print "unique values in type: " + str(df_tiny.TYPE.unique())


#append state household data to each row (so we can see how this census tract compared to the entire state)

df_tiny = df_tiny.merge(df_states, on='StateCode',indicator=True)
df_tiny['IncomeLevel'] = df_tiny['MHI_2010_x'] / df_tiny['MHI_2010_y']

#create qct, family, dummy and other dummy variables
df_tiny= pd.get_dummies(df_tiny, prefix='TargetPop', columns=['TRGT_FAM'])
df_tiny= pd.get_dummies(df_tiny, prefix='ConsType', columns=['TYPE'])
df_tiny= pd.get_dummies(df_tiny, prefix='Metro', columns=['METRO'])
df_tiny= pd.get_dummies(df_tiny, prefix='DDA', columns=['DDA'])
df_tiny= pd.get_dummies(df_tiny, prefix='QCT', columns=['QCT'])

```

```python
#create distribution of income percentages
sns.distplot(df_tiny['IncomeLevel'], kde=False, rug=True,bins=50)
df_tiny['IncomeLevel'].describe()
```

```python
# Any census tract with income at or above its state median income can be considered "high income".

#I've noted it with the category "income threshold" below. This would put it in the 75th percentile and above for all Census Tracts
# with LIHTC projects
df_tiny['IncomeThresh'] = 0
df_tiny.loc[df_tiny.IncomeLevel > 1, 'IncomeThresh'] = 1

#I ended up not using this!
```

```python
#remove outliers in the data set
#50% growth rate is probably unrealistic and is a data error in the ACS Median Income data.
df_tiny = df_tiny[(df_tiny.PercentGrowth_x < .5)]

```

### Aggregate census tracts with multiple properties on them to sum the total numbers of units. 

```python
#group by census tract. use this to sum the total number of units in a single census tract (so we don't repeat)
df_group = df_tiny.groupby('TRACT_LEVE').agg(sum)
df_group = df_group.reset_index()
df_group = df_group[['TRACT_LEVE','N_UNITS','LI_UNITS']]

#now do a merge based on df_tiny
df = pd.merge(df_tiny, df_group, on='TRACT_LEVE', how='inner')
```

## Training the model

### Run the model
I thought an SGD regressor would have been better but it turns out linear_model.Linear Regression works very quickly and is probably more accurate.

Linear regression is simple. I at first thought I'd have to add polynomials or log tranform my data, but it turns out I didn't have to due to my R-squred value being so high.

The get_linear_metrics gives me a lot of good data, including MSEs and P-values.

```python
#here we go!
X = df[['MHI_2010_x','LI_UNITS_x','TargetPop_1','DDA_1']]
#'IncomeThresh' I decided to throw away the comparison to the state Income due to colilnearity above. 
y = df['MHI_2014_x']

#i could square the variables for a polynomial fit. or log transform. doesn't seem neccessary.

#lm = linear_model.SGDRegressor(random_state=4)  #Try SGD to see if it works with a constant random state.
lm = linear_model.LinearRegression()
lm.fit(X, y)

print "~~~~Linear Model Metrics~~~~~"
lm = get_linear_model_metrics(X, y, lm)
print 'MSE', metrics.mean_squared_error(y, lm.predict(X))
```

#### Lasso and Ridge
Lasso and ridge wont solve high bias, will only help for variance.
My R-squareds are actually pretty good so I don't know if these will help. But it was worth running these anyway


```python
lm = linear_model.Lasso().fit(X, y)
print "~~~ Lasso ~~~"
print 'Lasso MSE: ', metrics.mean_squared_error(y, lm.predict(X))
print 'Lasso R2:', lm.score(X, y)


lm = linear_model.Ridge().fit(X, y)
print "~~~ Ridge ~~~"
print 'Ridge MSE: ', metrics.mean_squared_error(y, lm.predict(X))
print 'Ridge R2:', lm.score(X, y)
```

### Cross Validating the Model
It appears the model is consistent when cross validating throuhg 10 folds. Thanks again for this handy code!

```python
kf = cross_validation.KFold(len(X), n_folds=10, shuffle=True)

mse_values = []
scores = []
n= 0
print "~~~~ CROSS VALIDATION each fold ~~~~"
for train_index, test_index in kf:
    lm = linear_model.LinearRegression().fit(X.iloc[train_index], y.iloc[train_index])
    mse_values.append(metrics.mean_squared_error(y.iloc[test_index], lm.predict(X.iloc[test_index])))
    scores.append(lm.score(X, y))
    print 'Model', n
    print 'MSE:', mse_values[n-1]
    print 'R2:', scores[n-1]


print "~~~~ SUMMARY OF CROSS VALIDATION ~~~~"
print 'Mean of MSE for all folds:', np.mean(mse_values)
print 'Mean of R2 for all folds:', np.mean(scores)
```
